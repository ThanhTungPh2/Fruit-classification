```python
import numpy as np
import pandas as pd
import math
from collections import Counter
import random
```

```python
data = pd.read_csv("./sample_data/fruit_data_with_colours.csv")
data = data.drop(["Unnamed: 7","Unnamed: 8", "Unnamed: 9"], 1)
data["mass"] = (data["mass"]/data["mass"].max()) * 100
data["width"] = (data["width"]/data["width"].max()) * 100
data["height"] = (data["height"]/data["height"].max()) * 100
data["color_score"] = (data["color_score"]/data["color_score"].max()) * 100
shuffle = data.sample(frac = 1)
shuffle
```

```python
trainSet = shuffle[1:55] #hehe
testSet = shuffle[55:]
print(trainSet)
print(testSet)
```

```python
def computeDistance(dataPoint1, dataPoint2):
  result = 0;
  for i in range(4):
    result += (float(dataPoint1[i+3]) - float(dataPoint2[i+3]))**2;
  return math.sqrt(result)
```

```python
def computeKNN(trainSet, item, k):
  distance = []
  for index, row in trainSet.iterrows():
    distance.append(
        {
            "label": row[1],
            "value": computeDistance(item, row)
        }
    )
  distance.sort(key = lambda x:x["value"])
  labels = [item["label"] for item in distance]
  return labels[:k]
```

```python
def voteTheDistantce(knn):
  label = Counter(knn).most_common()
  return label
```

```python
k = 3
for item,row in testSet.iterrows():
  knn = computeKNN(trainSet, row, k)
  result = voteTheDistantce(knn)
  print("label: {} --> predic {}".format(row[1],result))
```
